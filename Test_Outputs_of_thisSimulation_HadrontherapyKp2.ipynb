```python
def wk0(n, p):
    base = (1 - 1/(2*n))
    exponent = (1 - 1/p)
    return (1 - pow( base, exponent))

def wkg(n, p, k):
    exponent = (1 - 1/p)
    base1 = 1 - (k - 1/2)/n
    base2 = 1 - (k + 1/2)/n
    return (pow(base1, exponent) - pow(base2, exponent))
            
def wkn(n, p):
    exponent = (1 - 1/p)
    return pow(1/(2*n), exponent)
```

```python
def plotSOBP(dataDirectory, nBeams, initEb, stepEb, pExp):
    """
    =======
    This formula is from the following paper:
     "Creating a spread-out Bragg peak in proton beams" by David Jette and Weimin Chen
     ======
    
    The model of Bortfeld and Schlegel (1996 Phys. Med. Biol. 41 1331–9) for
    determining the weights of proton beams required to create a spread-out Bragg
    peak (SOBP) ....
    
    The work is based on a close power-law relationship between the range R of 
    protons and their energy E:
        R = αE^{p0} , (1)
    where α = 0.0022 and p0 = 1.77. The values of these two parameters are properly taken from
    a subsequent study (Bortfeld 1997) which develops analytic incorporation of range straggling
    into the Bragg curve.
    
    The SOBP is divided into n equal intervals, and monoenergetic beams of range equal to 
    the depth of the ends of these intervals are used, with weights given by wk for 
    k = 0, 1, . . . , n. 
    """
    #n = 10 #15 #9
    n = nBeams - 1  # Number of divisions of SOBP = One less than the # of pristine peaks or Ebeams
    
    #p = 1.73 #1.77 #1.425 #2.2 #1.9
    p = pExp
    
    ######### Next creating the list of weights ############
    """
    # Following is what I used to do for n = 10 (without using loop)
    wt = [wk0(n,p), wkg(n,p,1), wkg(n,p,2), wkg(n,p,3), wkg(n,p,4), 
           wkg(n,p,5), wkg(n,p,6), wkg(n,p,7), wkg(n,p,8), wkg(n,p,9), 
           wkn(n,p)]
    """ 
    #Starting the weight list with the one for the first pure Bragg peak
    wt = [wk0(n,p)] # list of weights, more will be appended in a loop

    # Appending all the others except the last one.
    for i in range (1,n):
        wt.append(wkg(n,p,i))
    
    wt.append(wkn(n,p)) # Weight for the last pristine Bragg peak

    print(wt)
    #######################

    Ei = initEb #170
    dE = stepEb #2
    dfArr = []
    dfWtArr = []
    file = ""

    #while (i < 9):  #works but needs the incrementer line i = i+1
    for i in range(0,nBeams): #11):
        #print(i)
        #file = "./Results/V1_10kEvs/Let_gpsPr{}.00MeV.out".format(Ei + i*dE)
        file = dataDirectory + "Let_gpsPr{}.00MeV.out".format(initEb + i*stepEb)
        df = pd.read_csv(file,delim_whitespace=True)
        dfArr.append(df)
        dfWtArr.append(wt[i]*df)
        #print(dfArr[i])
        print("Appended dataframe of " + file + " to the dataframe Dict.")

        
    ################# Now lets draw the plots #########
    rcParams['figure.figsize'] = 16 , 16 #kp: without this line, default figure size was too small

    #fig, axes = plt.subplots(2, 2, subplot_kw=dict(polar=True))
    fig, axes = plt.subplots(2, 2)#, subplot_kw=dict(polar=True))

    #colName = "Dose(Gy)" # Only for Dose???.out file
    colName = "Dose"
    #colName = "LDT"
    #colName = "LTT"

    for i in range(0,nBeams):
        axes[0, 0].plot(dfArr[i][colName])
        axes[0, 1].plot(wt[i]*dfArr[i][colName])

    #frames180to220 = [df180, df185, df190, df195, df200, df205, df210, df215, df220]
    #resSum180to220 = pd.concat(frames180to220).groupby(level=0).sum()
    simpleSum = pd.concat(dfArr).groupby(level=0).sum()
    weightedSum = pd.concat(dfWtArr).groupby(level=0).sum()

    #axes[1, 1].scatter(x, y)
    axes[1, 0].plot(simpleSum[colName], color="Green")
    axes[1, 1].plot(weightedSum[colName], color="Green")
    axes[0, 0].title.set_text('Dose (Pristine Peaks - unscaled)')
    axes[0, 1].title.set_text('Dose (Pristine Peaks - scaled)')
    axes[1, 0].title.set_text('Unweighted Sum')
    axes[1, 1].title.set_text('SOBP - weighted sum')
  
```
